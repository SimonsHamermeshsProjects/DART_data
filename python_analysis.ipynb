```python
import pandas as pd
import folium
import numpy as np
path = 'csv/'
```

```python
data = pd.read_csv(path+'DART_data.csv')
```

```python
route_num=2
route = data[data['route_id']==route_num]
```

```python
route = route.drop_duplicates('stop_id')
```

```python
route = route.sort('stop_id')
```

```python
bus_map = folium.Map(location=[39.7458, -75.5467],zoom_start=10)
```

```python
prev = [0,0]
for index, row in route.iterrows():
    folium.Marker([row['latitude'], row['longitude']], popup=row['stop_name']).add_to(bus_map)
    # Would be nice to add lines, but for now this is a mess
    #if index != 0:
    #    current = (row['latitude'],row['longitude'])
    #    folium.features.PolyLine(locations=[current,prev]).add_to(bus_map)
    #    prev = current
```
